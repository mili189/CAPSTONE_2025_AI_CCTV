## **2. ëª¨ë¸ ë° ì˜ìƒ ë¶ˆëŸ¬ì˜¤ê¸°**

```python

# í•™ìŠµëœ YOLO ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
MODEL_PATH = "/content/drive/MyDrive/Capstone_Code/YOLO Predict/best.pt"

if not os.path.exists(MODEL_PATH):
    print(f"ğŸš¨ ëª¨ë¸ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {MODEL_PATH}")
    print("ì¢Œì¸¡ íŒŒì¼ íƒ­ì—ì„œ 'runs/detect/train/weights/best.pt' ê²½ë¡œë¥¼ í™•ì¸í•´ì£¼ì„¸ìš”.")
else:
    model = YOLO(MODEL_PATH)
    print(f"âœ… ëª¨ë¸ ë¡œë“œ ì™„ë£Œ: {MODEL_PATH}")

#  ì˜ˆì¸¡ë  ì˜ìƒ ê²½ë¡œ ì„¤ì •
SOURCE_VIDEO = "/content/drive/MyDrive/Capstone_Code/YOLO Predict/00151_H_A_BY_C1.mp4"
OUTPUT_VIDEO = "/content/drive/MyDrive/Capstone_Code/YOLO Predict/predict_video/result_video.mp4"
```

## **3. ì˜ˆì¸¡ ì‹œì‘**

```python
# ì˜ìƒ íŒŒì¼ ì—´ê¸°
cap = cv2.VideoCapture(SOURCE_VIDEO)

# ì˜ìƒ ì •ë³´ ì½ê¸° (ë„ˆë¹„, ë†’ì´, FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# ë¹„ë””ì˜¤ ì €ì¥ ì„¤ì • (ì½”ë±: mp4v)
video_writer = cv2.VideoWriter(
    OUTPUT_VIDEO,
    cv2.VideoWriter_fourcc(*'mp4v'),
    fps,
    (w, h)
)

print(f"ğŸš€ ë¶„ì„ ì‹œì‘... (ì´ {total_frames} í”„ë ˆì„)")

frame_count = 0

while cap.isOpened():
    success, frame = cap.read()

    if success:
        frame_count += 1

        # í”„ë ˆì„ ë‹¨ìœ„ë¡œ ì˜ˆì¸¡
        results = model.track(frame, persist=True, verbose=False)

        # ê²°ê³¼ ì‹œê°í™” (ë°”ìš´ë”© ë°•ìŠ¤ ê·¸ë¦¬ê¸°)
        annotated_frame = results[0].plot()

        # ì˜ìƒ íŒŒì¼ì— ì“°ê¸°
        video_writer.write(annotated_frame)

        # ì§„í–‰ ìƒí™© ì¶œë ¥ (50í”„ë ˆì„ë§ˆë‹¤)
        if frame_count % 50 == 0:
            print(f"ğŸ”¥ ì§„í–‰ ì¤‘: {frame_count}/{total_frames} ({(frame_count/total_frames)*100:.1f}%)")

    else:
        break

# ì¢…ë£Œ í›„ ìì› í•´ì œ
cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
